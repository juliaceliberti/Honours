Trying a histogram to plot the distribution of gene expression levels

```python
import pandas as pd
import plotly.express as px

# Load RNA expression data
rna_expression_df = pd.read_csv("../rna_expression_list.csv")

# Apply the ceiling
rna_expression_df["expression"] = rna_expression_df["expression"].clip(upper=1000)

# Sort the dataframe by expression values for better visualization
rna_expression_df_sorted = rna_expression_df.sort_values(by="expression", ascending=False)

# Create the plot with Plotly
fig = px.bar(rna_expression_df, x=rna_expression_df.index, y='expression', title='RNA Expression Levels Across Genes', color_discrete_sequence=['rgba(0, 0, 255, 1)'] )

# Update layout for better readability
fig.update_layout(
    xaxis_title='Gene Index',
    yaxis_title='Expected Count of RNA Expression',
    height=600,  # Set the height of the plot
    width=1500 # Set the width of the plot
    
)

fig.show()
```

exploring class sizes for different thresholds

```python
gene_data_classes = gene_data.copy()
```

```python
gene_data_classes['thr0'] =  (gene_data_classes["expression"].values > 0).astype(int)
display(gene_data_classes)
print(gene_data_classes['thr0'].value_counts())
```

```python
gene_data_classes['thr1'] =  (gene_data_classes["expression"].values > 1).astype(int)
display(gene_data_classes)
print(gene_data_classes['thr1'].value_counts())
```

```python
gene_data_classes['thr3'] =  (gene_data_classes["expression"].values > 3).astype(int)
display(gene_data_classes)
print(gene_data_classes['thr3'].value_counts())
```

```python
gene_data_classes['thr5'] =  (gene_data_classes["expression"].values > 5).astype(int)
display(gene_data_classes)
print(gene_data_classes['thr5'].value_counts())
```
